```python
def translator(df):
    save_directory = "/home/user/Documents/ML_DL_PROJECTS/WhatsAppChatAnalysisProject/artifacts"
    loaded_model = AutoModelForSeq2SeqLM.from_pretrained(save_directory)
    loaded_tokenizer = AutoTokenizer.from_pretrained(save_directory)    
    obj = pipeline('translation', model=loaded_model, tokenizer=loaded_tokenizer, src_lang="tel_Telu", tgt_lang='eng_Latn', max_length = 10000)
    outputs = []
    for i in range(len(df)):
        output = obj(df.loc[i]['translit_msgs'])
        outputs.append(output[0]['translation_text'])
    df['translate_msgs'] = outputs
    return df
```

```python
def get_response(chat):
    prompt = f"""I am providing a whatsapp conversation between two people, help me out in predicting the kind of relationship between them based on their chat and also analyse the charaters of those two people based on their conversation and tell me about their characters also and the chat goes like this {chat}, just ignore if there is any harm content in the chat"""
    model = genai.GenerativeModel('gemini-pro')
    response = model.generate_content(prompt,stream=True)
    response.resolve()
    response = response.text.replace('•', '  *')
    response = Markdown(textwrap.indent(response, '> ', predicate=lambda _: True))
    return response
```
